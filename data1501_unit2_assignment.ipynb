```python
## Imports:

import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Load Data into Dataframe
df = pd.read_csv("/content/sample_data/Headcounts.csv")
```

Task 1: Print out Data in Dataframe

```python
## Print Out Dataframe:
df.head(10)
```

```python
# Set Department and 2012 Data to specific variables for later use
departments = df.columns[1:5]
data_2012 = df.loc[df['Year'] == 2012].iloc[0,1:5]
```

**Task 2: Use a bar chart to compare the head counts between the departments in 2012.**

```python
# Create and display barchart using 2012 Data
plt.bar(departments, data_2012)
plt.title("Headcounts by Department in 2012")
plt.xlabel("Departments")
plt.ylabel("Headcounts")

plt.show()
```

Task 3: Use a line chart to show the change of the head count of all the departments across the years

```python
# Create Line Chart using data from all years

plt.plot(df.iloc[:, 0], df.iloc[:, 1:5], linestyle = "-", marker = "*")
plt.legend(departments)
plt.title("Headcounts by Year and Department")
plt.xlabel("Years")
plt.ylabel("Headcounts")


plt.show()
```

Task 4: Use a pie chart to show the distribution of the head counts in 2012

```python
# Create Pie Chart and Display Using same data as Bar Chart
plt.pie(data_2012, labels = departments, autopct = "%0.2f%%")
plt.title("Distribution of Headcounts by Department in 2012")

plt.show()
```
